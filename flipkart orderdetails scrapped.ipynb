```python
#import all libraries
import selenium
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup as bs
import requests
import urllib.request as ul
import pandas as pd
import time
#intialise chrome driver manager
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.flipkart.com/")
#using users account details and opening flipkart using selenium
# putting username and password of user in paranthesis
USERNAME = ''
PASSWORD = ''
user_id = driver.find_element_by_xpath("/html/body/div[2]/div/div/div/div/div[2]/div/form/div[1]/input")

user_id.send_keys(USERNAME)
passc =driver.find_element_by_xpath("/html/body/div[2]/div/div/div/div/div[2]/div/form/div[2]/input")
passc.send_keys(PASSWORD)
#opens the users account by clicking login button
driver.find_element_by_xpath("/html/body/div[2]/div/div/div/div/div[2]/div/form/div[3]/button").click()
```

```python
# opening order details page of user where all orders are present 
order = driver.get('https://www.flipkart.com/account/orders')
order

```

```python
# as all the orders not present in single page so we need to scroll down to last so that all orders will present is single page
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
# creating empty list and dictionary to store details
details = []
temp = {}
# list order contain list of all orders
listorder = []
# scrapping starts from here
# a common class is found for all the orders 
mainclass = driver.find_elements_by_class_name('ShLswe')

# looped through each order link to get the property

for i in mainclass:
    # class for indivisual order
    subsequentclass = i.find_elements_by_class_name('_2WFi0x')
    
    for j in subsequentclass :
        
        
        link = j.get_property('href')
        listorder.append(link)
        
        # time gap to load each page
        time.sleep(2)
   
    time.sleep(2)
    
    



# get into individual order page
for k in listorder:
    # openining each individual link
        driver.get(k)
        time.sleep(4)
        try:
            # finding order date of each orders if it is present else will give 'n/a'
            orderdate = driver.find_element_by_class_name("_3wS5ZT").text
        except:
            orderdate = 'n/a'
            continue
        try:
            # finding product name
            name = driver.find_element_by_class_name("col-8-12").text
        except:
            name = 'n/a'
            continue
        try:
            # finding amount of the product
            amount = driver.find_element_by_class_name("_13fZDJ").text
        except:
            amount = 'n/a'
            continue
        try:
            # if order is not cancelled then packed date will extracted 
            packeddate = driver.find_elements_by_class_name("_3wS5ZT")[1].text
        except:
            packeddate = 'n/a'
            continue
        try:
            # finding shipped date . if not present then will give 'n/a'
        
            shippeddate = driver.find_elements_by_class_name("_3wS5ZT")[2].text
        except:
            shippeddate = 'n/a'
            continue
        try:
            # finding deleivery date of product if order is confirmed
            deleiverdate = driver.find_elements_by_class_name("_3wS5ZT")[3].text
        except:
            deleiverdate = 'n/a'
            continue
        # class for product link so that we can extract category of order
        elementdetail = driver.find_element_by_class_name("col-8-12")
        try:
            element = elementdetail.find_element_by_tag_name('a')
            element2 = element.get_property('href')
            driver.get(element2)
        except:
            continue
       
        time.sleep(2)
        try:
            # extracting order category
            ordercategory = driver.find_elements_by_class_name("_1KHd47")[1].text
        except:
            ordercategory = 'n/a'
        # saving all the details of each order in dictionary with individual name
        temp ={'name of product':name,'amount':amount,'packed date':packeddate,'shipped date':shippeddate,'deleivered date': deleiverdate,'ordercategory':ordercategory}
        # appending all the details in list 
        details.append(temp)
        
       
    
# further coming back to order details page of user       
order = driver.get('https://www.flipkart.com/account/orders')
order
time.sleep(5)
# using pandas table for order details was created
pd.DataFrame(details)

    
 
```
